```python
import pandas as pd
import os
import matplotlib.pyplot as plt
folder = "I:/Research Issues/Data/Al/"
files = filter(lambda a:'.xls' in a, os.listdir(folder))
# for f in files:
#     print f
s = pd.read_excel(folder + files[0])

SUMs = {}
for i,f in enumerate(files):
    s = pd.read_excel(folder + f)
    SUMs['{0}'.format(1992+i)]=s.SUM
    
stock = pd.DataFrame(SUMs)
stock['Name']=s.Name
```

```python
%matplotlib inline

for i in xrange(18):
    city = stock.loc[i]
    city.shape
    type(city[:19])
    plt.plot(range(1992,2011),city[:19])
    plt.show()
```
