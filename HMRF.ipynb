```python
# load modules
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import numpy as np
from PIL import Image
from osgeo import ogr
from osgeo import osr
from osgeo import gdal
import os
```

```python
%matplotlib inline
# data source direvtory
def foo(a):
    def bar(t):
        if t==True:
            return 1
        else:
            return 0
    return map(bar,a)
dnbDir='I:/DATA/VIIRS/DNBClip2014/'
ccDir='I:/DATA/VIIRS/CloudCoverClip20142015/'
dnbs=map(lambda a: dnbDir+a,filter(lambda b:b[-3:] == 'tif',os.listdir(dnbDir)))
dnb=dnbs[0]
dnbTif=gdal.Open(dnb)
band=dnbTif.GetRasterBand(1)
raster = band.ReadAsArray(0,0).astype(np.float)

judge=raster>band.GetNoDataValue()
raster1=map(foo,judge)*raster
raster1=raster1/raster1.max()*255
im=plt.imshow(raster1,cmap=cm.terrain)
plt.show()
```
